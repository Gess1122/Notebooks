### TAF 2 : Statistiques descriptives

- 1. Calculer les mesures de tendance centrale (moyenne, médiane) et de dispersion (écart-type,
variance) pour les variables Age, AnnualIncome et SpendingScore.


```python
# calculons les moyennes
moy_age = df['Age'].mean()
moy_ann = df['AnnualIncome'].mean()
moy_sp = df['SpendingScore'].mean()

print("La moyenne de la variable age: ", moy_age)
print("La moyenne de la variable AnnuallIncome: ", moy_ann)
print("La moyenne de la variable SpendingScore: ", moy_sp)
print("\n")

# calculons les medianes
med_age = df['Age'].median()
med_ann = df['AnnualIncome'].median()
med_sp = df['SpendingScore'].median()
print("La mediane de la variable age: ", med_age)
print("La mediane de la variable AnnuallIncome: ", med_ann)
print("La mediane de la variable SpendingScore: ", med_sp)
print("\n")

# calculons l'ecart-type
ec_age = df['Age'].std()
ec_ann = df['AnnualIncome'].std()
ec_sp = df['SpendingScore'].std()
print("L'ecart-type de la variable age: ", ec_age)
print("L'ecart-type de la variable AnnuallIncome: ", ec_ann)
print("L'ecart-type de la variable SpendingScore: ", ec_sp)
print("\n")

# calculons la variance
var_age = df['Age'].var()
var_ann = df['AnnualIncome'].var()
var_sp = df['SpendingScore'].var()
print("La variance de la variable age: ", var_age)
print("La variance de la variable AnnuallIncome: ", var_ann)
print("La variance de la variable SpendingScore: ", var_sp)
```

- 2. Identifier les valeurs minimales et maximales de chaque variable.

```python
# les minimums
min_age = df['Age'].min()
min_ann = df['AnnualIncome'].min()
min_sp = df['SpendingScore'].min()
min_mn = df['MembershipDuration'].min()
print("Le minimum de la variable age: ", min_age)
print("Le minimum de la variable AnnuallIncome: ", min_ann)
print("Le minimum de la variable SpendingScore: ", min_sp)
print("Le minimum de la variable MembershipDuration : ", min_mn)
print("\n")

# les maximums
max_age = df['Age'].max()
max_ann = df['AnnualIncome'].max()
max_sp = df['SpendingScore'].max()
max_mn = df['MembershipDuration'].max()
print("Le minimum de la variable age: ", max_age)
print("Le minimum de la variable AnnuallIncome: ", max_ann)
print("Le minimum de la variable SpendingScore: ", max_sp)
print("Le minimum de la variable MembershipDuration : ", max_mn)
print("\n")

```

### TAF 5 : Identification des motifs et tendances

- 1. Examiner les clusters naturels dans les données à l’aide d’un scatterplot des variables
AnnualIncome et SpendingScore.

```python
from sklearn.cluster import kMeans
kmeans = kMeans(n_clusters=3, random_state=42)
df['cluster'] = kmeans.fit_predict(df[['AnnualIncome', 'SpendingScore']])
plt.figure(figsize=(8, 6))
plt.scatter(df['AnnualIncome'], df['SpendingScore'], c=df['cluster'], cmap='viridis', edgecolor='k')
plt.title("AnnualIncome et SpendingScore")
plt.xlabel('AnnualIncome')
plt.ylabel('SpendingScore')
plt.grid(True)
plt.show()
```
